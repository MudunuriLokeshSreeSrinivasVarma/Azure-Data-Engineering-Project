```python
tables = [
"agg_doctor_performance",
"agg_patient_billing_summary",
"agg_revenue_by_month",
"agg_treatment_summary",
"dim_doctors",
"dim_patients",
"fact_appointments"
]

for table in tables:

    df = spark.read.format("delta").load(
    f"abfss://gold@hospitaldatalake1.dfs.core.windows.net/Hospital_data/{table}"
    )

    df.write \
    .format("jdbc") \
    .mode("overwrite") \
    .option("url", jdbcUrl) \
    .option("dbtable", table) \
    .options(**connectionProperties) \
    .save()

    print(f"{table} loaded to Azure SQL")
```
